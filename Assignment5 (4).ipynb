```python
!pip install -q --upgrade --force-reinstall langchain langchain-community langchain-core langchain-text-splitters langchain-huggingface pypdf faiss-cpu sentence-transformers transformers accelerate torch
from langchain.chains.retrieval import RetrievalQA
from langchain_core.prompts import PromptTemplate
from transformers import LlamaTokenizer, LlamaForCausalLM, pipeline

llm = pipeline(
    "text-generation",
    model="mistralai/Mistral-7B-Instruct-v0.2",
    device_map="auto",
    max_new_tokens=512,
    temperature=0.1
)

prompt = PromptTemplate(
    input_variables=["context", "question"],
    template="""
You are an insurance policy expert.

Answer ONLY from the policy text below.
If the answer is not explicitly stated, say "Not found in policy."

Include:
- Clear answer
- Supporting clause text
- Page number

Policy Text:
{context}

Question:
{question}
"""
)

qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=retriever,
    chain_type="stuff",
    return_source_documents=True,
    chain_type_kwargs={"prompt": prompt}
)
```
