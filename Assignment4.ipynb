Write a Python program to draw a 3D plot that visualizes the regression model for house price prediction using suitable Python-based 3D plotting libraries.

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from mpl_toolkits.mplot3d import Axes3D

# -----------------------------
# Step 1: Create Sample Dataset
# -----------------------------

# Feature 1: Area in square feet
area = np.array([800, 1000, 1200, 1500, 1800, 2000]).reshape(-1, 1)

# Feature 2: Number of bedrooms
bedrooms = np.array([1, 2, 2, 3, 3, 4]).reshape(-1, 1)

# Target: House Price (in thousands)
price = np.array([120, 180, 200, 280, 320, 360])

# Combine features
X = np.hstack((area, bedrooms))
y = price

# --------------------------------
# Step 2: Train Regression Model
# --------------------------------

model = LinearRegression()
model.fit(X, y)

# --------------------------------
# Step 3: Create Grid for 3D Plane
# --------------------------------

area_range = np.linspace(area.min(), area.max(), 20)
bedroom_range = np.linspace(bedrooms.min(), bedrooms.max(), 20)

area_grid, bedroom_grid = np.meshgrid(area_range, bedroom_range)

# Predict prices on grid
price_pred = model.predict(
    np.c_[area_grid.ravel(), bedroom_grid.ravel()]
).reshape(area_grid.shape)

# -----------------------------
# Step 4: 3D Visualization
# -----------------------------

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot (actual data)
ax.scatter(area, bedrooms, price,
           color='red', s=50, label='Actual Prices')

# Surface plot (regression plane)
ax.plot_surface(area_grid, bedroom_grid, price_pred,
                alpha=0.6, cmap='viridis')

# Labels
ax.set_xlabel('Area (sq ft)')
ax.set_ylabel('Number of Bedrooms')
ax.set_zlabel('House Price (in thousands)')
ax.set_title('3D Regression Model for House Price Prediction')

ax.legend()

plt.show()
```

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from mpl_toolkits.mplot3d import Axes3D

# -----------------------------
# Step 1: Create Sample Dataset
# -----------------------------

# Feature 1: Area in square feet
area = np.array([800, 1000, 1200, 1500, 1800, 2000]).reshape(-1, 1)

# Feature 2: Number of bedrooms
bedrooms = np.array([1, 2, 2, 3, 3, 4]).reshape(-1, 1)

# Target: House Price (in thousands)
price = np.array([120, 180, 200, 280, 320, 360])

# Combine features
X = np.hstack((area, bedrooms))
y = price

# --------------------------------
# Step 2: Train Regression Model
# --------------------------------

model = LinearRegression()
model.fit(X, y)

# --------------------------------
# Step 3: Create Grid for 3D Plane
# --------------------------------

area_range = np.linspace(area.min(), area.max(), 10)
bedroom_range = np.linspace(bedrooms.min(), bedrooms.max(), 10)

area_grid, bedroom_grid = np.meshgrid(area_range, bedroom_range)

# Predict prices on grid
price_pred = model.predict(
    np.c_[area_grid.ravel(), bedroom_grid.ravel()]
).reshape(area_grid.shape)

# -----------------------------
# Step 4: 3D Visualization
# -----------------------------

fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot (actual data)
ax.scatter(area, bedrooms, price,
           color='green', s=50, label='Actual Prices')

# Surface plot (regression plane)
ax.plot_surface(area_grid, bedroom_grid, price_pred,
                alpha=0.1, cmap='viridis')

# Labels
ax.set_xlabel('Area (sq ft)')
ax.set_ylabel('Number of Bedrooms')
ax.set_zlabel('House Price (in thousands)')
ax.set_title('3D Regression Model for House Price Prediction')

ax.legend()

plt.show()
```
